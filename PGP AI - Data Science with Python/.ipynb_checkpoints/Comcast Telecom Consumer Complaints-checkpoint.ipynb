#### Importing Data in Python

```python
data = pd.read_csv('datasets/Comcast_telecom_complaints_data.csv')
```

#### Displaying the top 5 rowsof the dataset

```python
data.head()
```

#### Creating a new column 'infer_date_time' using Pandas to_datetime method
*Using the info method we can see the new column added as a datetime type*

```python
data['infer_date_time'] = pd.to_datetime(data['Date_month_year']+' '+data['Time'])
data.info()
```

#### Analysing Year and Month from infer_date_time column
*Year is found to be 2015 and month is numericaly indexed*

```python
year = pd.DatetimeIndex(data['infer_date_time']).year
year.unique()
```

```python
month = pd.DatetimeIndex(data['infer_date_time']).month
month.unique()
```

#### Extracting the Month from infer_date_time into a new column 'complaint_month' to perform aggregate functions

```python
data['complaint_month'] = pd.DatetimeIndex(data['infer_date_time']).month
data.info()
```

#### Displaying the distribution of complaints across 12 months in bar chart
*June Month witnessed highest complaints*

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

complaint_status_by_month = data.groupby(['complaint_month'])['complaint_month'].count().fillna(0)

plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(15,6))
plt.bar(complaint_status_by_month.index, complaint_status_by_month.values)

plt.ylabel('Complaint Counts')
plt.title('Month wise Complaint Status')
plt.xticks(complaint_status_by_month.index, months)

plt.show()
```

```python
data['complaint_hour'] = pd.DatetimeIndex(data['infer_date_time']).hour
data.info()
```

```python
complaint_by_hour = data.groupby('complaint_hour')['complaint_hour'].count().fillna(0)
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(15,6))
plt.plot(complaint_by_hour.index, complaint_by_hour.values)

plt.ylabel('Complaint Counts')
plt.title('Hour wise Complaints')
plt.xticks(complaint_by_hour.index)
plt.show()
```

```python
data['Status'].value_counts()
```

```python
data['complaint_status'] = data['Status'].apply(lambda x : 'Open' if x in ['Open', 'Pending'] else 'Closed').astype('category')
data.info()
```

```python
data['complaint_status'].value_counts()
```

```python
complaint_status_by_state = data.groupby('State')['State'].count().fillna(0)
complaint_status_by_state.values
```

```python
complaint_status_by_state = data.groupby(['State', 'complaint_status'])['State'].count().unstack('complaint_status').fillna(0)
```

```python
complaint_status_by_state[['Closed','Open']].plot(kind='bar', stacked=True, figsize=(20,8), legend=True, 
                                                  title='State wise Complaint Status')
plt.style.use('ggplot')
plt.show()
```

```python
uresolved_percentage_array = complaint_status_by_state['Open'] / \
                       (complaint_status_by_state['Open'] + \
                       complaint_status_by_state['Closed']) * 100
pd.DataFrame(uresolved_percentage_array, columns=['uresolved_percentage']).sort_values(by=['uresolved_percentage']) \
                                                    .plot(kind='barh', figsize=(15,10), yticks=[0,100,10], 
                                                          title='State wise Unresolve Percentage')
plt.style.use('seaborn')
plt.show()
```

```python
customer_complaints = data['Customer Complaint']

stop_words = {'ourselves', 'hers', 'between', 'yourself', 'but', 'again', 'there', 'about', 'once', 'during', 'out', 
              'very', 'having', 'with', 'they', 'own', 'an', 'be', 'some', 'for', 'do', 'its', 'yours', 'such', 'into', 
              'of', 'most', 'itself', 'other', 'off', 'is', 's', 'am', 'or', 'who', 'as', 'from', 'him', 'each', 'the', 
              'themselves', 'until', 'below', 'are', 'we', 'these', 'your', 'his', 'through', 'don', 'nor', 'me', 'were', 
              'her', 'more', 'himself', 'this', 'down', 'should', 'our', 'their', 'while', 'above', 'both', 'up', 'to', 
              'ours', 'had', 'she', 'all', 'no', 'when', 'at', 'any', 'before', 'them', 'same', 'and', 'been', 'have', 'in', 
              'will', 'on', 'does', 'yourselves', 'then', 'that', 'because', 'what', 'over', 'why', 'so', 'can', 'did', 'not', 
              'now', 'under', 'he', 'you', 'herself', 'has', 'just', 'where', 'too', 'only', 'myself', 'which', 'those', 'i', 
              'after', 'few', 'whom', 't', 'being', 'if', 'theirs', 'my', 'against', 'a', 'by', 'doing', 'it', 'how', 
              'further', 'was', 'here', 'than'}

complaint_words = []
for complaint in customer_complaints:
    words = complaint.split(' ')
    for word in words:
        if word.lower() not in stop_words:
            complaint_words.append([word.upper(), len(word)])
len(complaint_words)
```

```python
df_complaint_words = pd.DataFrame(complaint_words, columns=['Word','Length'])
df_complaint_words = df_complaint_words.drop_duplicates()
df_complaint_words['Length'].describe()
```

```python
df_complaint_words = df_complaint_words.drop(df_complaint_words[df_complaint_words['Length'] <= 4].index)
df_complaint_words['Length'].describe()
```
