```python
import sys
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
print('Python Version: {}'.format(sys.version))
print('Numpy Version: {}'.format(np.__version__))
print('Pandas Version: {}'.format(pd.__version__))
print('Matplotlib Version: {}'.format(plt.__version__))
```

```python
data = pd.read_csv('datasets/Comcast_telecom_complaints_data.csv')
```

```python
data.info()
```

```python
data.head()
```

```python
data['infer_date_time'] = pd.to_datetime(data['Date_month_year']+' '+data['Time'])
data.info()
```

```python
data.head(10)
```

```python
data['Status'].value_counts()
```

```python
data['complaint_status'] = data['Status'].apply(lambda x : 'Open' if x in ['Open', 'Pending'] else 'Closed').astype('category')
data.info()
```

```python
data['complaint_status'].value_counts()
```

```python
complaint_status_by_state = data.groupby(['State', 'complaint_status'])
complaint_status_by_state.agg('count').head(10)
```

```python
complaint_status_by_state = data.groupby(['State', 'complaint_status'])['State'].count().unstack('complaint_status').fillna(0)
complaint_status_by_state[['Closed','Open']].plot(kind='bar', stacked=True, figsize=(20,8), legend=True, 
                                                  title='State wise Complaint Status')
plt.show()
```

```python
year = pd.DatetimeIndex(data['infer_date_time']).year
year.unique()
```

```python
month = pd.DatetimeIndex(data['infer_date_time']).month
month.unique()
```

```python
data['complaint_month'] = pd.DatetimeIndex(data['infer_date_time']).month
data.info()
```

```python
complaint_status_by_month = data.groupby(['complaint_month', 
                                          'complaint_status'])['complaint_month'].count().unstack('complaint_status').fillna(0)
complaint_status_by_month[['Closed','Open']].plot(kind='bar', stacked=True, figsize=(20,8), legend=True, 
                                                  title='Month wise Complaint Status')
plt.show()
```
